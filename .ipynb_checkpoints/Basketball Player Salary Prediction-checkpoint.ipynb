```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv("nba2k20-full.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df=df.dropna()
```

```python
df.isnull().sum().sum()
```

```python
df=df.drop(['full_name','b_day','height','weight','college'],axis=1)
df.head()
```

```python
def removehash(value):
    value=value[1:]
    return int(value)
df['jersey']=df['jersey'].apply(removehash)
```

```python
df.head()
```

```python
#Data Cleaning
df['salary']=df['salary'].apply(removehash)
df.head()
```

```python
df['team'].value_counts()  #unique values in the team
```

```python
df['country'].value_counts()
```

```python
def removecountryoutlier(value):
    if value not in ['USA','Canada','Australia']:
        return 'Others'
    else:
        return value
df['country']=df['country'].apply(removecountryoutlier)
df['country'].value_counts()
```

```python
df['position'].value_counts()
```

```python
df.head()
```

```python
df['draft_round'].unique
```

```python
df['draft_peak'].unique
```

```python
def removedrafted(value):
    if value=='Undrafted':
        return
    else:
        return value
df['draft_round']=df['draft_round'].apply(removedrafted)
df['draft_peak']=df['draft_peak'].apply(removedrafted)
df=df.dropna()
    
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['position']=le.fit_transform(df['position'])
df['country']=le.fit_transform(df['country'])
df['team']=le.fit_transform(df['team'])
df.head()
```

```python
df.describe()
```

```python
x=df.drop(['salary'],axis=1)
y=df['salary']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train, y_train)
predictions=lr.predict(x_test)
```

```python
diff=y_test - predictions
```

```python
import seaborn as sns
sns.histplot(diff)
```

```python
import pickle
pickle.dump(lr, open('./model.sav','wb'))
```
